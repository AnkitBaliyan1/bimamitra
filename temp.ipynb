```python


# Create an empty list to store the repeated rows
repeated_rows_list = []

# Iterate through each row in the original DataFrame
for _, row in df.iterrows():
    # Check for NaN values in "CSC members" and replace with 0
    csc_members = int(row['CSC members']) if not np.isnan(row['CSC members']) else 0
    # Repeat the current row based on the value in "CSC members"
    repeated_rows = [row] * csc_members
    repeated_rows_list.extend(repeated_rows)

# Create a new DataFrame from the list of repeated rows
new_df = pd.DataFrame(repeated_rows_list)

# Reset the index of the new DataFrame
new_df.reset_index(drop=True, inplace=True)


```
